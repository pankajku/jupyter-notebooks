```python
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

import pandas as pd
from datetime import datetime

init_notebook_mode(connected=True)
def price_chart(tsymbols, investment = 10000, sdate = None, edate = None, ptype = 'Adj Close'):
    dfs = []
    for tsymbol in tsymbols:
        df = pd.read_csv('./{}.csv'.format(tsymbol))
        if sdate:
            df = df.loc[df['Date'] >= sdate]
        if edate:
            df = df.loc[df['Date'] <= edate]
        df = df.loc[:, ['Date', ptype]]
        
        num_shares = investment/df.at[df.first_valid_index(), ptype]
        col = '{} NAV'.format(tsymbol)
        df[col] = num_shares * df[ptype]
        dfs.append({'df': df, 'col': col})

    iplot([{'x': df['df'].Date, 'y': df['df'][df['col']],'name': df['col']} for df in dfs], filename = 'time-series-simple')

def price_chart2(tsymbols, investment = 10000, sdate = None, edate = None):
    dfs = []
    for tsymbol in tsymbols:
        df = pd.read_csv('./{}.csv'.format(tsymbol))
        if sdate:
            df = df.loc[df['Date'] >= sdate]
        if edate:
            df = df.loc[df['Date'] <= edate]
            
        col = '{}'.format(tsymbol)
        df1 = df.loc[:, ['Date', 'Close']]
        num_shares = investment/df1.at[df1.first_valid_index(), 'Close']
        df1[col] = num_shares * df1['Close']
        dfs.append({'df': df1, 'col': col})

        col = '{} NAV'.format(tsymbol)
        df2 = df.loc[:, ['Date', 'Adj Close']]
        num_shares = investment/df2.at[df2.first_valid_index(), 'Adj Close']
        df2[col] = num_shares * df2['Adj Close']
        dfs.append({'df': df2, 'col': col})
        
    iplot([{'x': df['df'].Date, 'y': df['df'][df['col']],'name': df['col']} for df in dfs], filename = 'time-series-simple')
```

```python
price_chart(['VWEHX', 'TGLMX', 'SPY'], investment= 10000, sdate='1995-01-01', edate='2019-10-01', ptype='Adj Close')
```
